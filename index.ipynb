#### 3.3 Recodificando variables 

```python
#Verificando el tipo de variables

tipos_variables = idh_distrital4.dtypes

print("Tipos de variables en el DataFrame:")
print(tipos_variables)
```

```python
idh_distrital5 = idh_distrital4.copy()
```

```python
#renombrando columnas
nuevos_nombres = {
    'UBIGEO': 'ubigeo',
    'Distrital' : 'distrital',
    'Población' : 'poblacion',
    'Índice de Desarrollo Humano' : 'idhumano',
    'Esperanza de vida al nacer' : 'esperanza_vida',
    'Con Educación secundaria completa (Poblac. 18 años)' : 'secundaria_completa',
    'Años de educación (Poblac. 25 y más)' : 'años_educacion',
    'Ingreso familiar per cápita' : 'ingreso_fam'
}

idh_distrital5.rename(columns = nuevos_nombres, inplace = True)
print(idh_distrital5)
```

```python
#recodificando el tipo de variables 
##variables float 

columnas_reco = [
    "idhumano",
    "esperanza_vida",
    "secundaria_completa",
    "años_educacion", 
    "ingreso_fam",   
]

idh_distrital5[columnas_reco] = idh_distrital5[columnas_reco].astype(float)

print("Tipos de variables después de la conversión:")
print(idh_distrital5.dtypes)
```

```python
##variables int 
columnas_reco2 = ["ubigeo", "poblacion"]

idh_distrital5[columnas_reco2] = idh_distrital5[columnas_reco2].astype(int)

print("Tipos de variables después de la conversión:")
print(idh_distrital5.dtypes)

##variables str

columnas_reco3 = ["distrital"]

idh_distrital5[columnas_reco3] = idh_distrital5[columnas_reco3].astype(str)

print("Tipos de variables después de la conversión:")
print(idh_distrital5.dtypes)
```

### 4. Filtrar la base de datos a nivel de distritos de la región Ica

```python
#Distritos del departamento de ICA = 110

patron = '^110' #patron de ubigeo que empieza con 110

idh_distrital5_ica = idh_distrital5[idh_distrital5['ubigeo'].astype(str).str.contains(patron)]

print(idh_distrital5_ica)
```

## Visualización de gráficos


### Gráfico 1: años de educación a nivel distrital 

```python
import altair as alt
import wbgapi as wb
```

```python
alt.Chart(idh_distrital5_ica).mark_point()
```

```python
alt.Chart(idh_distrital5_ica).mark_point().encode(
    x = "distrital")
```

```python
graf1 = alt.Chart(idh_distrital5_ica).mark_point().encode(
    y=alt.Y("años_educacion", title="Años de Educación"),
    x=alt.X("distrital", title="Distritos"),
    color=alt.Color("distrital", title="Distritos"),
).properties(
    title="Años de educación a nivel distrital en la región de Ica",
).interactive()

graf1 = graf1.configure_legend(
    labelFontSize=10,
    titleFontSize=11,
)

graf1
```

### Gráfico 2: comparación entre indicadores de IDH (Años de educación y Ingreso familiar per cápita)

```python
graf2 = alt.Chart(idh_distrital5_ica).mark_point().encode(
    y = alt.Y("años_educacion:Q", title= "Años de Educación"),
    x = alt.X("ingreso_fam:Q", title="Ingreso familiar per cápita"),
    color=alt.Color("ingreso_fam", title="Ingreso familiar per cápita"),
).properties(
    title="Relación entre el ingreso familiar per cápita y los años de educación a nivel distrital en Ica",
    ).interactive()

graf2
```
